```
#Question 3
#How to convert the index of a series into a column of a dataframe?

import pandas as pd
import numpy as np


sa = pd.Series(["Mumbai", "Delhi", "Chennai", "Jaipur", "Benguluru"])

df = pd.DataFrame(sa, columns=['City']) # Converting series to df with columns name
#df.reset_index(level=0, inplace=True) 
df['index'] = df.index    # series index as df column
df

```
